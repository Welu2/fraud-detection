```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

credit = pd.read_csv(
    "../data/processed/creditcard_processed.csv"
)
```

```python
credit.head()
credit["Class"].value_counts(normalize=True)*100
```

```python

sns.countplot(
    x="Class",
    data=credit
)
```

```python
sns.histplot(
    credit["Amount"],
    kde=True
)
```

```python
corr = credit.corr()



plt.figure(figsize=(15,10))

sns.heatmap(
    corr,
    cmap="coolwarm"
)
```
